A landscape image.  Note, the image must be a "*.png".

```julia
using PyPlot
using SVDapprox
SVDimage = download("http://files.myopera.com/salmanbabri/albums/7457652/Scenery.png", "svdImage.png");
img=float64(imread(SVDimage))
imshow(img)
```

Color images have three planes, for the red, green and blue colors respectively.

```julia
imshow(hcat([img[:,:,i] for i=1:3]...), cmap="hot")
```

We can calculate the SVD of each color plane at various rank approximations.  White colors don't always work that well.

```julia
# Function for the approximate SVD
function img_approx(img,k::Integer)
   imgout = zeros(size(img));
   (Ur,Sr,Vr) = svd_approx(img[:,:,1],k);
   (Ug,Sg,Vg) = svd_approx(img[:,:,2],k);
   (Ub,Sb,Vb) = svd_approx(img[:,:,3],k);
   imgout[:,:,1] = Ur*diagm(Sr)*Vr';
   imgout[:,:,2] = Ug*diagm(Sg)*Vg';
   imgout[:,:,3] = Ub*diagm(Sb)*Vb';
   return imgout
end

# Function for the true SVD
function img_true(img,k::Integer)
   imgout = zeros(size(img));
   (Ur,Sr,Vr) = svd_approx(img[:,:,1],k);
   (Ug,Sg,Vg) = svd_approx(img[:,:,2],k);
   (Ub,Sb,Vb) = svd_approx(img[:,:,3],k);
   imgout[:,:,1] = Ur[:,1:k]*diagm(Sr[1:k])*Vr[:,1:k]';
   imgout[:,:,2] = Ug[:,1:k]*diagm(Sg[1:k])*Vg[:,1:k]';
   imgout[:,:,3] = Ub[:,1:k]*diagm(Sb[1:k])*Vb[:,1:k]';
   return imgout
end

# Ranks to use
k_val = hcat([10,20,40],
             [60,80,100],
             [125,150,200])
Kx,Ky = size(k_val)

# Get the SVD estimates
svd_est  = vcat([hcat([img_approx(img,k_val[i,j]) for i=1:Kx]...) for j=1:Ky]...)
svd_true = vcat([hcat([img_true(img,k_val[i,j]) for i=1:Kx]...) for j=1:Ky]...)

# Plot
fig1 = figure()
ax1 = subplot(1,1,1)
imshow(svd_est)
setp(ax1[:get_xticklabels](),visible=false)
setp(ax1[:get_yticklabels](),visible=false)
savefig("svd_est.png")

fig2 = figure()
ax2 = subplot(1,1,1)
imshow(svd_true)
setp(ax2[:get_xticklabels](),visible=false)
setp(ax2[:get_yticklabels](),visible=false)
savefig("svd_true.png")

fig3 = figure()
(U,Sr,V) = svd(img[:,:,1])
(U,Sg,V) = svd(img[:,:,2])
(U,Sb,V) = svd(img[:,:,3])
xvals = [1:100]
y1 = Sr[1:100]
y2 = Sg[1:100]
y3 = Sb[1:100]
p1 = plot(xvals,y1,"ro-")
p2 = plot(xvals,y2,"go-")
p3 = plot(xvals,y3,"bo-")
xlab = xlabel(L"$i$")
ylab = ylabel(L"$\sigma_i$")
setp(p1,"linewidth",4,"MarkerSize",6)
setp(p2,"linewidth",4,"MarkerSize",6)
setp(p3,"linewidth",4,"MarkerSize",6)
legend([p1,p2,p3],["Red Channel","Green Channel","Blue Channel"])
setp(xlab,"FontSize",20)
setp(ylab,"FontSize",20)
savefig("svd_vals.png")
```
